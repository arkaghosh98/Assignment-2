```python
dummy_list = [3, 5, 2, 5, 9]
```

```python
print(dummy_list)
```

```python
dummy_list.reverse()
print(dummy_list)
```

```python
dummy_list_2 = [2, 200, 16, 4, 1, 0, 9.45, 45.67, 90, 12.01, 12.02]
dummy_list += dummy_list_2
print(dummy_list)
```

```python
dummy_dict = {}
for e in dummy_list:
  if e not in dummy_dict:
    dummy_dict[e] = 1
  else:
    dummy_dict[e] += 1
```

```python
print(dummy_dict)
```

```python
dummy_list.sort()
print(dummy_list)

dummy_list.sort(reverse = True)
print(dummy_list)
```

```python
x = 200

dummy_list.remove(x)
print(dummy_list)
```
